```python
import hashlib
import time

class BasitGuvenliRNG:
    def __init__(self, anahtar_kelime):
        # Başlangıç sayısını (Seed) oluşturuyoruz
        # Kullanıcı metni + o anki zamanın nanosaniyesi
        seed_hazirlik = anahtar_kelime + str(time.time_ns())
        self.seed = int(hashlib.sha256(seed_hazirlik.encode()).hexdigest(), 16)

        # LCG Parametreleri (Standart değerler)
        self.m = 2**32  # Sayıların üst sınırı
        self.a = 1664525
        self.c = 1013904223

    def sayi_uret(self):
        # 1. Adım: Matematiksel üretim (LCG)
        self.seed = (self.a * self.seed + self.c) % self.m

        # 2. Adım: Çıktıyı Hashle (Güvenlik Katmanı)
        # Bu adım, sayının bir sonraki tahmini zorlaştırır
        hash_obj = hashlib.sha256(str(self.seed).encode())
        hex_cikti = hash_obj.hexdigest()

        # Hex çıktısını anlaşılır bir tam sayıya çeviriyoruz (0-1000 arası)
        return int(hex_cikti, 16) % 1000

# --- GÖRSEL VE METİNSEL ÇIKTI ---

def raporla():
    anahtar = input("Rastgelelik için bir kelime girin: ")
    rng = BasitGuvenliRNG(anahtar)

    print("\n--- ÜRETİM RAPORU ---")
    print(f"Başlangıç Durumu: {anahtar} sistemi kuruldu.\n")

    sayilar = []
    for i in range(1, 11):
        sayi = rng.sayi_uret()
        sayilar.append(sayi)
        # Görselleştirme: Sayı büyüklüğüne göre yıldız basma
        grafik = "*" * (sayi // 25)
        print(f"Sayı {i:02d}: {sayi:4d} | {grafik}")

    print("\n--- İSTATİSTİKSEL ANALİZ ---")
    print(f"En Büyük Sayı: {max(sayilar)}")
    print(f"En Küçük Sayı: {min(sayilar)}")
    print(f"Ortalama Değer: {sum(sayilar)/len(sayilar)}")

if __name__ == "__main__":
    raporla()
```
